```python
dfBigAV = df.nlargest(n=35, columns= "average_rating")
dfBigAV
```

```python
dfBigAV.plot.scatter(x="authors", y="average_rating")
plt.xticks(rotation=90)
plt.xlabel('Authors')
plt.ylabel('Average Rating')
plt.title('Average Rating for the Top 35 Authors')
```

```python
dfBigAV.plot.bar(x="publisher", y="average_rating",legend=None)
plt.xticks(rotation=90)
plt.xlabel('Publisher')
plt.ylabel('Average Rating')
plt.title('Average Rating for the Top 35 Publishers')
#Plot of Average Rating for the Top 35 Publishers
```
